# Q9) histogram of critics rating

# Q11) show the total count of all the different genre present in the dataset 

```python
movie_data['Genre'].value_counts()
```

# Q12) show the graph of rating based on genre 

```python
sns.jointplot(movie_data['Genre'],movie_data['Audience Ratings %'])

plt.show()
```

# Q13) scatter plot of critics rating vs audience rating 

```python
plt.scatter(x='Rotten Tomatoes Ratings %',y='Audience Ratings %', data=movie_data, color='r')
plt.xlabel('critic rating')
plt.ylabel('audience rating')
plt.title('scatter plot')
plt.show()
```

# Q14) box plot based on genre and critics rating

# Q15) box plot based on genre and audience rating
